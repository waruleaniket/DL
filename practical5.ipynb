```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pylab as pylab
import numpy as np
```

```python
import re
```

```python
sentences = """The forest is home to thousands of plants and animals.
Rivers flow through the valleys providing water to the land.
Birds migrate across continents in search of food and better climate.
Mountains stand tall and protect the region from strong winds.
Nature balances life and every creature has an important role to play."""
```

```python
sentences = re.sub('[^A-Za-z0-9]+', ' ', sentences)
sentences = re.sub(r'(?:^| )\w(?:$| )', ' ', sentences).strip()
sentences = sentences.lower()
```

```python
words = sentences.split()
vocab = set(words)
```

```python
vocab_len = len(vocab)
embed_dim = 10
context_size = 2
```

```python
word_to_ix = {}
ix_to_word = {}

i = 0
for word in vocab:
    word_to_ix[word] = i
    ix_to_word[i] = word
    i += 1
```

```python
data = []
for i in range(2, len(words) - 2):
    context = [words[i-2], words[i-1], words[i+1], words[i+2]]
    target = words[i]
    data.append((context, target))
print(data[:5])
```

```python
embeddings = np.random.random_sample((vocab_len, embed_dim))
```

```python
#hidden â†’ output layer
def linear(m, theta):
    w = theta
    return m.dot(w)
```

```python
def log_softmax(x):
    e_x = np.exp(x - np.max(x)) # softmax to prevent overflow
    return np.log(e_x / e_x.sum()) # 
```

```python
def NLLLoss(logs, targets):
    out = logs[range(len(targets)), targets]
    return -out.sum()/len(out)
```

```python
def log_softmax_crossentropy_with_logits(logits,target):

    out = np.zeros_like(logits)
    out[np.arange(len(logits)),target] = 1
    
    softmax = np.exp(logits) / np.exp(logits).sum(axis=-1,keepdims=True)
    
    return (- out + softmax) / logits.shape[0]
```

```python
def forward(context_idxs, theta):
    m = embeddings[context_idxs].reshape(1, -1)
    n = linear(m, theta)
    o = log_softmax(n)
    
    return m, n, o
```

```python
def backward(preds, theta, target_idxs):
    m, n, o = preds
    
    dlog = log_softmax_crossentropy_with_logits(n, target_idxs)
    dw = m.T.dot(dlog)
    
    return dw
```

```python
def optimize(theta, grad, lr=0.03):
    theta -= grad * lr
    return theta
```

```python
theta = np.random.uniform(-1, 1, (2 * context_size * embed_dim, vocab_len))
```

```python
epoch_losses = {}

for epoch in range(80):

    losses =  []

    for context, target in data:
        context_idxs = np.array([word_to_ix[w] for w in context])
        preds = forward(context_idxs, theta)

        target_idxs = np.array([word_to_ix[target]])
        loss = NLLLoss(preds[-1], target_idxs)

        losses.append(loss)

        grad = backward(preds, theta, target_idxs)
        theta = optimize(theta, grad, lr=0.03)
        
     
    epoch_losses[epoch] = losses
```

```python
ix = np.arange(0,80)

fig = plt.figure()
fig.suptitle('Epoch/Losses', fontsize=20)
plt.plot(ix,[epoch_losses[i][0] for i in ix])
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Losses', fontsize=12)
```

Predict Function

```python
def predict(words):
    context_idxs = np.array([word_to_ix[w] for w in words])
    preds = forward(context_idxs, theta)
    word = ix_to_word[np.argmax(preds[-1])]
    
    return word
```

```python
# (['we', 'are', 'to', 'study'], 'about')
predict(['is', 'home' ,'to' ,'thousands'])
```

```python
def accuracy():
    wrong = 0

    for context, target in data:
        if(predict(context) != target):
            wrong += 1
            
    return (1 - (wrong / len(data)))
```
